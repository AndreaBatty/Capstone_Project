Plot the distribution curve for the column `ConvertedComp`.

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(x=df['ConvertedComp'])
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)
range_inter = q3 - q1
print(range_inter)
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = q1 - 1.5 * range_inter
upper_bound = q3 + 1.5 * range_inter

print(lower_bound)
print(upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
top_outliers = df[df['ConvertedComp'] > upper_bound]
bottom_outliers = df[df['ConvertedComp'] < lower_bound]

print(f"Numero di outliers inferiori: {bottom_outliers.shape[0]}")
print(f"Numero di outliers superiori: {top_outliers.shape[0]}")
print(f"Numero totale di outliers: {(top_outliers-bottom_outliers).shape[0]}")
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df2 = df[df['ConvertedComp'] <= upper_bound]
print(df2['ConvertedComp'].median())
print(df2['ConvertedComp'].mean())
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# your code goes here
# Seleziona solo colonne numeriche
import numpy as np
numeric_df = df.select_dtypes(include=[np.number])

# Calcola la correlazione
correlation = numeric_df.corr()
print(correlation['Age'])
```
